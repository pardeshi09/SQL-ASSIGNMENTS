### Q-19. Write an SQL query to determine the 5th highest salary without using TOP or limit method.

SELECT * FROM (SELECT ROW_NUMBER() OVER (ORDER BY SALARY DSC) AS R_NUM, FIRST_NAME, SALARY DROM employee ) AS HIGHEST_SALARY WHERE R_NUM = 5
